```python
#top 10 most affected country
top_10_countries = df.groupby('Country/Region')['Confirmed'].sum().sort_values(ascending=False).head(10)
```

```python
top_10_countries
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x  = 'confirmed', y = 'Country/Region', data = 'top_10_countries')
plt.xlabel('Confirmed')
plt.ylabel('Country/Region')
plt.title('Top 10 Most Affected Countries')
plt.show()
```

```python
# top 10 countries with highest death rate

top_10_countries_death = df.groupby('Country/Region')['Deaths / 100 Cases'].sum().sort_values(ascending=False).head(10)
```

```python
top_10_countries_death
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Deaths / 100 Cases', y='Country/Region', data = top_10_countries_death)
plt.xlabel('Deaths / 100 Cases')
plt.ylabel('Country/Region')
plt.title('Top 10 Most Affected Countries')
plt.show()
```

```python
top_10_countries_death = top_10_countries_death.reset_index()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x='Deaths / 100 Cases', y='Country/Region', data = top_10_countries_death)
plt.xlabel('Deaths / 100 Cases')
plt.ylabel('Country/Region')
plt.title('Top 10 Most Affected Countries')
plt.show()
```
